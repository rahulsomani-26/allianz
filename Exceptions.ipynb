```python
class InputMissingError(AttributeError):
    """When you miss the input"""
```

```python
try:
    raise InputMissingError("Please fill in the required field")
except Exception as e:
    print(e.__class__,e.args)
```

```python
issubclass(InputMissingError,AttributeError)
```

```python
issubclass(InputMissingError,Exception)
```

```python
issubclass(InputMissingError,BaseException)
```

```python
issubclass(InputMissingError,int)
```
