```python
from bs4 import BeautifulSoup 
import requests
response = requests.get('https://automatetheboringstuff.com/')
if response.status_code==200:
    soup = BeautifulSoup(response.text,'html.parser')
    title = soup.find_all('h2')
```

```python
for items in title:
    print(items.text)
```

```python
from bs4 import BeautifulSoup 
import requests
response = requests.get('https://www.google.com/search?q=gold+price')
soup = BeautifulSoup(response.text,'html.parser')
price = soup.find_all('span')
```

```python
for items in price:
    print(items.text)
```
